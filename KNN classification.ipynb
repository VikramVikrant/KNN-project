### To predict the class of the car by feeding unlabelled data & using KNN Algorithm

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model,preprocessing
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv('D:\Data_file\kNN\Car Data Set\car.data')
```

```python
df.head() #maint:maintenance
```

```python
#Converting all our string data to numeric form
obj=preprocessing.LabelEncoder()
buying=obj.fit_transform(list(df['buying']))
maint=obj.fit_transform(list(df['maint']))
door=obj.fit_transform(list(df['door']))
persons=obj.fit_transform(list(df['persons']))
lug_boot=obj.fit_transform(list(df['lug_boot']))
safety=obj.fit_transform(list(df['safety']))
cls=obj.fit_transform(list(df['class']))
```

```python
print(buying) #it has created it in numpy array with integers value
```

```python
print(maint)
```

```python
#we are going tp predict the class
predict='class'

X=list(zip(buying,maint,door,persons,lug_boot,safety))
y=list(cls)

x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(X,y,test_size=0.2)
```

```python
#print(x_test)
```

```python
print(y_test)
```

```python
model=KNeighborsClassifier(n_neighbors=6)

model.fit(x_train,y_train)
accuracy=model.score(x_test,y_test)
print(accuracy)
```

```python
best=0
for n_neighbors in range(1,10):
    model=KNeighborsClassifier(n_neighbors)
    model.fit(x_train,y_train)
    accuracy=model.score(x_test,y_test)
    if accuracy>best:
        best=accuracy
    print(f'the value of k is {n_neighbors} & is with accuracy {best}.')    
```

```python
# importing all scoring methods
from sklearn.metrics import jaccard_similarity_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
from sklearn.model_selection import cross_val_score
```

```python
bestScore=0.0
accuracyList=[]

for k in range(3,12):
    
    clf_knn = KNeighborsClassifier(n_neighbors=k,algorithm='auto')#classifier_knn
    
    # using 10 fold cross validation for scoring the classifier's accuracy
    scores = cross_val_score(clf_knn, X, y, cv=10)
    score=scores.mean()
    accuracyList.append(score)
    
    if score > bestScore:
        bestScore=score
        best_clf=clf_knn
        bestK=k
        
print("Best K is :",bestK,"| Cross validation Accuracy :",bestScore)
clf_knn=best_clf
```

```python
clf_knn.fit(x_train,y_train)
y_pred=best_clf.predict(x_train)
```

```python
##Keeping all scores of the classifiers in trainScores dictionary
trainScores={}
```

```python
trainScores['KNN-jaccard']=jaccard_similarity_score(y_train, y_pred)
trainScores['KNN-f1-score']=f1_score(y_train, y_pred, average='weighted')
```
